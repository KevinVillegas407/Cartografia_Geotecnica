Teniendo ya esto, podemos comenzar con el método

***Frecuency Ratio Model***

En este, se combinan el inventario de MenM con las diferentes variables, para así obtener el peso de cada clase
que componen las variables.

Recordemos que las variables elegidas fueron: Pendiente, Aspectos, Curvatura, Flujo Acumulado y Elevación.

```python
inventarioNanVcMenM = inventariovectorMenM.copy()
inventarioNanVcMenM[inventarioNanVcMenM == 0] = np.nan
```

**Pendientes**

```python
pendiente_serie = pd.Series(slopevectorMenM)
p1W = p2W = p3W = p4W = p5W = 0

for celda in pendiente_serie: ## Area de la clase
    if (celda >= 0) & (celda < 9):
        p1W += 1
    elif (celda >= 9) & (celda < 19):
        p2W += 1
    elif (celda >= 19) & (celda < 29):
        p3W += 1
    elif (celda >= 29) & (celda < 36):
        p4W += 1
    elif celda >= 36:
        p5W += 1

total = p1W + p2W + p3W + p4W + p5W ## Area total de las clases
print('Total Celdas:', total, p1W, p2W, p3W, p4W, p5W)

##### 

pendienteConMenM = slopevectorMenM * inventarioNanVcMenM
p1W_1 = p2W_1 = p3W_1 = p4W_1 = p5W_1 = 0

for celda in pendienteConMenM: ## MenM en las clases
    if (celda >= 0) & (celda < 9):
        p1W_1 += 1
    elif (celda >= 9) & (celda < 19):
        p2W_1 += 1
    elif (celda >= 19) & (celda < 29):
        p3W_1 += 1
    elif (celda >= 29) & (celda < 36):
        p4W_1 += 1
    elif celda >= 36:
        p5W_1 += 1

total1 = p1W_1 + p2W_1 + p3W_1 + p4W_1 + p5W_1 ## MenM total en las clases
print('Total MenM:', total1, p1W_1, p2W_1, p3W_1, p4W_1, p5W_1)

##### 

p1WW = (p1W_1/total1)/(p1W/total)
p2WW = (p2W_1/total1)/(p2W/total)
p3WW = (p3W_1/total1)/(p3W/total)
p4WW = (p4W_1/total1)/(p4W/total)
p5WW = (p5W_1/total1)/(p5W/total)

print('')
print('Peso Clases:', p1WW, p2WW, p3WW, p4WW, p5WW)
print('')

### ---------------- Mapa con peso de las clases

pendienteFR = np.where ( (np.logical_and (slope >= 0, slope < 9 )), p1WW, slope );
pendienteFR = np.where ( (np.logical_and (pendienteFR >= 9, pendienteFR < 19)), p2WW, pendienteFR);
pendienteFR = np.where ( (np.logical_and (pendienteFR >= 19, pendienteFR < 29 )), p3WW, pendienteFR);
pendienteFR = np.where ( (np.logical_and (pendienteFR >= 29, pendienteFR < 36 )), p4WW, pendienteFR);
pendienteFR = np.where ( pendienteFR >= 36, p5WW, pendienteFR);


#plt.figure(figsize=(40, 10))  # Ajusta el tamaño
plt.imshow(pendienteFR)

#plt.savefig('PesoClasePendientes.png', dpi=500, bbox_inches='tight')
plt.colorbar();
```

**Aspectos**

```python
aspecto_serie = pd.Series(aspectovectorMenM)
asp1W = asp2W = asp3W = 0

for celda in aspecto_serie: ## Area de la clase
    if celda <= 125:
        asp1W += 1
    elif (celda >= 125) & (celda < 230):
        asp2W += 1
    elif celda >= 230:
        asp3W += 1

total = asp1W + asp2W + asp3W ## Area total de las clases
print('Total Celdas:', total, asp1W, asp2W, asp3W)

##### 

aspectoConMenM = aspectovectorMenM * inventarioNanVcMenM
asp1W_1 = asp2W_1 = asp3W_1 = 0

for celda in aspectoConMenM: ## MenM en las clases
    if celda <= 125:
        asp1W_1 += 1
    elif (celda >= 125) & (celda < 230):
        asp2W_1 += 1
    elif celda >= 230:
        asp3W_1 += 1

total1 = asp1W_1 + asp2W_1 + asp3W_1 ## MenM total en las clases
print('Total MenM:', total1, asp1W_1, asp2W_1, asp3W_1)

####

asp1WW = (asp1W_1/total1)/(asp1W/total)
asp2WW = (asp2W_1/total1)/(asp2W/total)
asp3WW = (asp3W_1/total1)/(asp3W/total)

print('')
print('Peso Clases:', asp1WW, asp2WW, asp3WW)
print('')

####
aspectoFR = np.where ( aspecto < 125, asp1WW, aspecto);
aspectoFR = np.where ( (np.logical_and (aspectoFR >= 125, aspectoFR < 230)), asp2WW, aspectoFR)
aspectoFR = np.where ( aspectoFR >= 230, asp3WW, aspectoFR);

#plt.figure(figsize=(40, 10))  # Ajusta el tamaño
plt.imshow(aspectoFR)
plt.colorbar();
```

**Curvatura**

```python
p40 = np.nanpercentile(curvatura1, 40)
p60 = np.nanpercentile(curvatura1, 60)

curvatura_serie = pd.Series(curvaturavectorMenM)
cur1W = cur2W = cur3W = 0

for celda in curvatura_serie: ## Area de la clase
    if (celda > p40) & (celda < p60):
        cur1W += 1
    elif celda >= p60:
        cur2W += 1
    elif celda <= p40:
        cur3W += 1

total = cur1W + cur2W + cur3W ## Area total de las clases
print('Total Celdas:', total, cur1W, cur2W, cur3W)

##### 

curvaturaConMenM = curvaturavectorMenM * inventarioNanVcMenM
cur1W_1 = cur2W_1 = cur3W_1 = 0

for celda in curvaturaConMenM: ## MenM en las clases
    if (celda > p40) & (celda < p60):
        cur1W_1 += 1
    elif celda >= p60:
        cur2W_1 += 1
    elif celda <= p40:
        cur3W_1 += 1

total1 = cur1W_1 + cur2W_1 + cur3W_1 ## MenM total en las clases
print('Total MenM:', total1, cur1W_1, cur2W_1, cur3W_1)

####

cur1WW = (cur1W_1/total1)/(cur1W/total)
cur2WW = (cur2W_1/total1)/(cur2W/total)
cur3WW = (cur3W_1/total1)/(cur3W/total)

print('')
print('Peso Clases:', cur1WW, cur2WW, cur3WW)
print('')

####
curvaturaFR = np.where ( (np.logical_and (curvatura1 > p40, curvatura1 <  p60)), cur1WW, curvatura1)
curvaturaFR = np.where(curvaturaFR >=  p60, cur2WW, curvaturaFR)
curvaturaFR = np.where(curvaturaFR <= p40, cur3WW, curvaturaFR)

#plt.figure(figsize=(40, 10))  # Ajusta el tamaño
plt.imshow(curvaturaFR)

#plt.savefig('PesoClaseCurvatura.png', dpi=500, bbox_inches='tight')
plt.colorbar();
```

**Flujo Acumulado**

```python
flujoAcum_serie = pd.Series(flujovectorMenM)
fac1W = fac2W = fac3W = 1

for celda in flujoAcum_serie: ## Area de la clase
    if (celda < 100):
        fac1W += 1
    elif (celda >= 100) & (celda <= 6000):
        fac2W += 1
    elif celda > 6000:
        fac3W += 1

total = fac1W + fac2W + fac3W ## Area total de las clases
print('Total Celdas:', total, fac1W, fac2W, fac3W)

#####

flujoAcumConMenM = flujovectorMenM * inventarioNanVcMenM
fac1W_1 = fac2W_1 = fac3W_1 = 1

for celda in flujoAcumConMenM: ## MenM en las clases
    if (celda < 100):
        fac1W_1 += 1
    elif (celda >= 100) & (celda <= 6000):
        fac2W_1 += 1
    elif celda > 6000:
        fac3W_1 += 1

total1 = fac1W_1 + fac3W_1 + fac2W_1 ## MenM total en las clases
print('Total MenM:', total1, fac1W_1, fac2W_1, fac3W_1)

#####

fac1WW = (fac1W_1/total1)/(fac1W/total)
fac2WW = (fac2W_1/total1)/(fac2W/total)
fac3WW = (fac3W_1/total1)/(fac3W/total)

print('')
print('Peso Clases:', cur1WW, cur2WW, cur3WW)
print('')

#####

flujoAcumFR = np.where ( flujo1 < 100, fac1WW, flujo1)
flujoAcumFR = np.where ( (np.logical_and (flujoAcumFR >= 100, flujoAcumFR < 6000)), fac2WW, flujoAcumFR)
flujoAcumFR = np.where ( flujoAcumFR > 6000, fac3WW, flujoAcumFR);

#plt.figure(figsize=(40, 10))  # Ajusta el tamaño
plt.imshow(flujoAcumFR)

#plt.savefig('PesoClaseFlujoAcumulado.png', dpi=500, bbox_inches='tight')
plt.colorbar();
```

**Elevación**

```python
elevacion_serie = pd.Series(elevacionvectorMenM)
al1W = al2W = al3W = al4W = al5W = al6W = 1

for celda in elevacion_serie: ## Area de la clase
    if (celda < 700):
        al1W += 1
    elif (celda >= 700) & (celda < 1000):
        al2W += 1
    elif (celda >= 1000) & (celda < 1700):
        al3W += 1
    elif (celda >= 1700) & (celda < 2100):
        al4W += 1
    elif (celda >= 2100) & (celda < 2500):
        al5W += 1
    elif celda >= 2500:
        al6W += 1

total = al1W + al2W + al3W + al4W + al5W ## Area total de las clases
print('Total Celdas:', total, al1W, al2W, al3W, al4W, al5W, al6W)

#####

elevacionConMenM = elevacionvectorMenM * inventarioNanVcMenM
al1W_1 = al2W_1 = al3W_1 = al4W_1 = al5W_1 = al6W_1 = 1

for celda in elevacionConMenM: ## MenM en las clases
    if (celda < 700):
        al1W += 1
    elif (celda >= 700) & (celda < 1000):
        al2W += 1
    elif (celda >= 1000) & (celda < 1700):
        al3W += 1
    elif (celda >= 1700) & (celda < 2100):
        al4W += 1
    elif (celda >= 2100) & (celda < 2500):
        al5W += 1
    elif celda >= 2500:
        al6W += 1

total1 = al1W_1 + al2W_1 + al3W_1 + al4W_1 + al5W_1 + al6W_1  ## MenM total en las clases
print('Total MenM:', total1, al1W_1, al2W_1, al3W_1, al4W_1, al5W_1, al6W_1)

#####

al1WW = (al1W_1/total1)/(al1W/total)
al2WW = (al2W_1/total1)/(al2W/total)
al3WW = (al3W_1/total1)/(al3W/total)
al4WW = (al4W_1/total1)/(al4W/total)
al5WW = (al5W_1/total1)/(al5W/total)
al6WW = (al6W_1/total1)/(al6W/total)


print('')
print('Peso Clases:', al1WW, al2WW, al3WW, al4WW, al5WW, al6W)
print('')

#####

altitudFR = np.where ( elevacion1 < 700, al1WW, elevacion1)
altitudFR = np.where ( (np.logical_and (altitudFR >= 700, altitudFR < 1000)), al2WW, altitudFR)
altitudFR = np.where ( (np.logical_and (altitudFR >= 1000, altitudFR < 1700)), al3WW, altitudFR)
altitudFR = np.where ( (np.logical_and (altitudFR >= 1700, altitudFR < 2100)), al4WW, altitudFR)
altitudFR = np.where ( (np.logical_and (altitudFR >= 2100, altitudFR < 2500)), al5WW, altitudFR)
altitudFR = np.where ( altitudFR >= 2500, al5WW, altitudFR)

#plt.figure(figsize=(40, 10))  # Ajusta el tamaño
plt.imshow(altitudFR)

#plt.savefig('MHeuristico_AltitudReclasificado.png', dpi=500, bbox_inches='tight')
plt.colorbar();
```

**APH**

```python
A = ([[  1, 3, 1/7, 1/5,   1, 1/3, 1],
      [1/3, 1, 1/7, 1/7, 1/3, 1/3, 1],
      [  7, 7,   1,   1,   1,   3, 7],
      [  5, 7,   1,   1,   3,   3, 5],
      [  1, 3,   1, 1/3,   1, 1/3, 1],
      [  3, 3, 1/3, 1/3,   3,   1, 7],
      [  1, 1, 1/7, 1/5,   1, 1/7, 1]])

B = np.array(A) # para transformarla en una matriz

valores, vectores = np.linalg.eig(B) # Función para calcular los valores y vectores propios

vector_norm = vectores/np.linalg.norm(vectores, ord=1) # normalización de los vectores

#print('Estos son los valores propios:\n',valores)
#print('\nEstos son los vectores propios:\n', vectores)
#print('\nEsto son los vectores propios normalizados:\n',vector_norm)

CI = (valores[0] - len(valores)) / (len(valores) - 1)

CR = CI / 1.32

if CR < 0.1:    print('La designación de datos fue aceptable:\n', CR)
else:
    print('La designación de datos NO fue aceptable:\n', CR)

# Para proseguir se utiliza el mayor vector propio, el que corresponde al de la posición 0
# El cual indicará los valores de los pesos de las variables introducidas

matrizIndices = ['slope', 'aspecto', 'curvatura', 'flujo acumulado', 'elevacion']
pesoVariables = vector_norm[:,0]
#print(pesoVariables)

for i in range(len(matrizIndices)):
    print(f'{matrizIndices[i]}:', pesoVariables[i])
```
